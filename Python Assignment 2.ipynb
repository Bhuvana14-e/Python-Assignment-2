**Sets (Operations):**

a. Create a set called my_set with following values:
['a','e','i','o','u','a','a','i']
Analyse the output and provide explanation for the same.

```python
my_set = {'a','e','i','o','u','a','a','i'}
print(my_set)
```

**Explanation:**

Sets store only unique values. Duplicate 'a' and 'i' are removed automatically, so the output will be: {'a', 'e', 'i', 'o', 'u'} .

b. Attempt to change the value of my_set[4] = 's'. If code throws an error, provide
an explanation.

```python
my_set[4] = 's'
```

**Explanation:**

Sets do not support indexing,Cannot access or modify elements by index. They are unordered and immutable in terms of positions.

c. Create two sets:

set1 with values: {1, 3, 5, 7, 9}
set2 with values: {2, 3, 5, 8, 10}

```python
set1 = {1, 3, 5, 7, 9}
set2 = {2, 3, 5, 8, 10}
print(set1)
print(set2)
```

d. Compute and print the union and intersection of set1 and set2.

```python
print("set union operation:",set1.union(set2))
print("set intersection operation:",set1.intersection(set2))
print("set intersection operation:",set2.intersection(set1))
```
